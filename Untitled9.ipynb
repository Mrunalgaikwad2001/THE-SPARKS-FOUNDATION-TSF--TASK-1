**Number of Matches in each season**

```python
num_matches_matches = matches.groupby('season')[['id']].count()
num_matches_matches
```

**Number of matches in each season**

```python
plt.title('Number of matches in each season',fontweight=1000)
plt.xlabel('Season')
plt.ylabel('Total no. of matches')
plt.xticks(num_matches_matches.index)
plt.bar(num_matches_matches.index,num_matches_matches.id,width=0.7, color=['red','yellow','blue','green','purple','grey',
'orange','pink','violet','brown','black','cyan'],alpha=0.8);
```

**Number of matches won by each team from 2008-2018**

```python
winner_matches = matches.groupby('winner')[['id']].count()
winner_matches = winner_matches.sort_values('id',ascending=False).reset_index()

winner_matches.rename(columns = {'id':'matches_won','winner':'team'},inplace=True)
winner_matches
```

**Number of matches in each season**

```python
plt.title('Number of matches won by each team from 2008-2018',fontweight=800)
plt.xlabel('Teams')
plt.ylabel('Total no.of matches won')
plt.xticks(rotation=90,fontsize=10)
plt.bar(winner_matches.team,winner_matches.matches_won,color=['red','yellow','blue','green','purple','grey',
'orange','pink','violet','brown','#158EA6','#E82865','#005DB7','#C23E25','#E82878'],alpha=0.8)
```

```python
final_match_matches = matches.groupby('season').tail(1).sort_values('season').reset_index()
final_match_matches
```

```python
plt.title('Count of Season winners',fontweight=800)
sns.countplot(x=final_match_matches['winner'],palette=['red','yellow','blue','green','purple','grey',]);
```

```python
final_match_matches['winner'].value_counts()
```

```python
final_match_team=pd.concat([final_match_matches['team1'],final_match_matches['team2']])
final_match_team.value_counts()
```

```python
plt.title('Count of finalists in IPL',fontweight=800)
plt.xlabel('Teams')
sns.countplot(x=final_match_team,palette=['#FFFF48','#1C2C46','#15244C','#CD202D','#D4480B','#ECC5F2','#EF2920','#F0E1A1','#E776CA'])
plt.ylabel('Final Matches Played')
plt.xticks(rotation=90);
```

```python
mom_count = matches.groupby('player_of_match')[['id']].count()
mom_count = mom_count.sort_values('id',ascending=False).head(10)
mom_count
```

```python
 plt.title("Top 10 player with Most Man of the Match Awards",fontweight=500)
 sns.barplot(x=mom_count.index,y=mom_count.id,alpha=0.6);
 plt.xticks(rotation=90)
 plt.yticks(ticks=np.arange(0,25,5))
 plt.ylabel('No of Awards')
 plt.xlabel('Players');
```

```python
plt.title("Top 10 player with Most Man of the Match Awards",fontweight=500)
sns.barplot(x=mom_count.index,y=mom_count.id,alpha=0.6);
plt.xticks(rotation=90)
plt.yticks(ticks=np.arange(0,25,5))
plt.ylabel('No of Awards')
plt.xlabel('Players')
```

```python
city_counts = matches.groupby('city').apply(lambda x:x['city'].count()).reset_index(name='Match Counts')
top_cities_order = city_counts.sort_values(by = 'Match Counts',ascending=False)
top_cities = top_cities_order[:10]
print('Top 10 Citizens with the maximum number of Matches Played:\n',top_cities)
plt.figure(figsize=(7,7))
plt.pie(top_cities['Match Counts'],labels=top_cities['city'],autopct='%1.1f%%',startangle=30)
plt.axis('equal')
plt.title('Top cities that have hosted IPL Matches',size=10)
```
